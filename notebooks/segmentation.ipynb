```python
import pandas as pd
from pandas import DataFrame
from pathlib import Path
from typing import List
from glob import glob
```

```python
def recursive_glob(root: str, pattern: str) -> List[Path]:
    return [Path(p) for p in glob(str(Path(root) / "**" / pattern), recursive=True)]
```

```python
data_directory = Path('/data/msc-proj')
train_dir = data_directory / 'Training_Set'
csv_ann_paths = recursive_glob(train_dir, '*.csv')
```

```python
def parse_csv_ann(csv_ann_path: Path) -> DataFrame:
    return pd.read_csv(csv_ann_path)
```
